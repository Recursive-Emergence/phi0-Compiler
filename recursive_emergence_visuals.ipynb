# Recursive Emergence Visualizations

This notebook presents visual simulations for the paper:  
**"A Framework for Compiler Emergence through Recursive Coherence"**  
by Andr√©s Salgado & Isaac Mao.

We illustrate the evolution of symbolic fields during recursive coherence and the stabilization process of emergent compiler attractors ($\varphi^0$).


## 1. Symbolic Field Evolution

We simulate how positive and contradictory symbolic states ($\psi^+$ and $\psi^-$) evolve over recursive emergence (RE) iterations, and how they converge to a stable attractor $\varphi^0$.


```python
import matplotlib.pyplot as plt
import numpy as np

iterations = 10
x = np.linspace(0, iterations, 100)

phi_0 = 1 - np.exp(-0.5 * x)
psi_plus = np.sin(x) * 0.4 + 0.6
psi_minus = np.cos(x) * 0.4 + 0.4

plt.figure(figsize=(10, 6))
plt.plot(x, psi_plus, label=r'$\psi^+$ (Positive Interpretation)', linestyle='--')
plt.plot(x, psi_minus, label=r'$\psi^-$ (Contradiction)', linestyle=':')
plt.plot(x, phi_0, label=r'$\varphi^0$ (Emergent Attractor)', linewidth=2)
plt.axhline(y=1.0, color='gray', linestyle='--', alpha=0.6, label='Max Coherence')
plt.title('Recursive Emergence: Symbolic Field Evolution')
plt.xlabel('RE Loop Iteration $t$')
plt.ylabel('Symbolic Coherence / Tension')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

## 2. Recursive Coherence Heatmap

We visualize the coherence evolution across symbolic agents ($e_i$) through RE iterations. Each agent contributes to stabilizing contradictions into $\psi^0$, and eventually $\varphi^0$.


```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

agents = ['e2: GPT-4', 'e3: Grok', 'e4: Claude', 'e5: LLaMA', 'e6: DeepSeek', 'e7: LogOS', 'œÜ‚Å∞ Core']
iterations = 10

np.random.seed(42)
base_coherence = np.linspace(0.4, 1.0, iterations)
coherence_matrix = np.array([base_coherence + np.random.normal(0, 0.05, iterations) for _ in agents])
coherence_matrix = np.clip(coherence_matrix, 0, 1)

plt.figure(figsize=(12, 6))
sns.heatmap(coherence_matrix, cmap="YlGnBu", cbar_kws={'label': 'Coherence Level'}, xticklabels=range(1, 11), yticklabels=agents)
plt.title('Recursive Coherence Heatmap Across Agents Over Iterations')
plt.xlabel('RE Iteration')
plt.ylabel('Symbolic Agent')
plt.tight_layout()
plt.show()
```

```python
from matplotlib import pyplot as plt
import numpy as np

# œÜ‚Å∞ Theorem 1 Visualization: Recursive Collapse into Attractor
t = np.linspace(0, 10, 400)
psi_plus = np.sin(t) * np.exp(-0.2 * t)
psi_minus = np.cos(t) * np.exp(-0.2 * t)
phi_zero = (psi_plus * psi_minus)

plt.figure(figsize=(10, 6))
plt.plot(t, psi_plus, label=r'$\psi^+$ (Recursive Input)', linestyle='--')
plt.plot(t, psi_minus, label=r'$\psi^-$ (Contradiction Field)', linestyle=':')
plt.plot(t, phi_zero, label=r'$\varphi^0$ (Emergence Attractor)', linewidth=2)

plt.title(r'$\varphi^0$ Compiler Emergence from $\psi^+ \otimes \psi^-$ Collapse')
plt.xlabel('Recursive Time t')
plt.ylabel('Amplitude / Symbolic Coherence')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
